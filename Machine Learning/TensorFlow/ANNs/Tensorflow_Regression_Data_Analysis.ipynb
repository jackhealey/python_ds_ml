## Feature engineering

```python
df.head()
```

```python
df.drop('id', axis = 1, inplace = True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].apply(lambda date : date.year)
df['month'] = df['date'].apply(lambda date : date.month)
```

```python
df.head()
```

```python
df.groupby('year')['price'].mean().plot()
```

```python
df.drop('date', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Is it viable to make 'zipcode' categorical?

df['zipcode'].value_counts()

# 70 zipcodes - therefore decided to drop
```

```python
df.drop('zipcode', axis = 1, inplace = True)
```

```python
df.head()
```

```python
# yr_renovated and sqft_basement: keep because 0 - highest should correlate with value.
```

## Data pre-processing

```python
X = df.drop('price', axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                X, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

## Creating a Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

# Base number of neurons in layers on size of feature data
# More layers risking overfitting
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

## Training the Model

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test),
                                                  batch_size = 128, epochs = 400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
# Could keep training without overfitting as val_loss follows loss down.
```

## Evaluation on Test Data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```
